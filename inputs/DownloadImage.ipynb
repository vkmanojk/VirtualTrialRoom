```python
import urllib
import json
```

```python
with open('FEIDEGGER.json') as json_data:
    d = json.load(json_data)
    idx=0
    for obj in d:
        url = json.dumps(obj['url'])
        print(url)
        testfile = urllib.URLopener()
        testfile.retrieve(url.replace('"',''), 'images/image'+str(idx)+'.jpg') 
        idx+=1
```
